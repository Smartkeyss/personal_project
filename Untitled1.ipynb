```python
import pandas as pd
from pandas import 
import numpy as np
```

```python
df = pd.read_csv("world_alcohol - world_alcohol.csv")
df
```

```python
#Write a Pandas program to filter those records where WHO region matches with multiple values (Africa, Eastern Mediterranean, Europe)
#from world alcohol consumption dataset
ls = ["Africa", "Eastern Mediterranean", "Europe"]
df[df["WHO region"].isin(ls)]
```

```python
#Write a Pandas program to filter those records which not appears in a given list from world alcohol consumption dataset
df[np.logical_not(df["WHO region"].isin(ls))]
```

```python
Write a Pandas program to #filter all records where the average consumption of beverages per person from .5 to 2.50
#in world alcohol consumption datase
df[df["Display Value"].all >= 0.5 or df["Display Value"].all <= 2.5]
```

```python
#Write a Pandas program to find number of consumption of wine per person greater than 2 in world alcohol consumption dataset
df[df["Display Value"] > 2].count()
```

```python
#Write a Pandas program to filter rows based on row numbers ended with
#0, like 0, 10, 20, 30 from world alcohol consumption dataset.
df[0::10]
```

```python
#Write a Pandas program to filter all records starting from the 2nd row,
#access every 5th row from world alcohol consumption dataset.
df[1::5]
```
