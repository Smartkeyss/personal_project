```python
import pandas as pd
import numpy as np
```

```python
#import the data
df = pd.read_csv("Complaints_Data.csv", )
```

```python
#Checking out the data
df.head()
```

```python
df = df.drop(0)
df.tail()
```

```python
df.tail()
```

```python
# checking data types of the data
df.dtypes
```

```python
#Checking unique values for the closed column
df["Closed"].unique()
```

```python
#Replace the null value with Open  because i felt that since it does not have a closing date, then it is still open  
df["Closed"].fillna("open", inplace= True)
```

```python
#Change Reported Data type to Date
df["Reported"] = pd.to_datetime(df["Reported"])
```

```python
# Group the Closed Column into open and Close
df["Open_or_Close"] = df["Closed"].apply(lambda x: "Open" if x == "open" else "Close")
```

```python
df.head()
```

```python
# How many are being reported closed and open
df.groupby("Open_or_Close").count()["Ref"]
```

```python
# How many are being reported on a monthly basis and how many are being closed (Overall)
df.groupby([df.Reported.dt.month, df.Open_or_Close]).count().unstack()["Ref"]
```

```python
# How many are being reported on a monthly basis and how many are being closed (Regional)
df.groupby([df["Region / Directorate"], df.Open_or_Close, df.Reported.dt.month]).count().unstack()["Ref"].head(22)
```

```python
# How many Stage 1s and 2are being reported and closed (overall)
df.groupby(["Current Stage", "Open_or_Close"]).count().unstack()["Ref"]
```

```python
#How many Stage 1s and 2s are being reported and closed (Regional)
df.groupby(["Current Stage", "Open_or_Close", "Region / Directorate"]).count().unstack()["Ref"]
```
